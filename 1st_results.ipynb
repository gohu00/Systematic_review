# Introduction

To have any hope of meeting the ambitious climate goals set by the Paris Agreement, we will have to implement an energy transition faster than originally planned (IEA, 2021). This imply to replace fossil energy sources with renewable energy, in the form of solar panels or wind turbines. The mineral resource requirements of renewables are becoming increasingly well known. The International Energy Agency (IEA) anticipates that to meet the needs of the energy transition alone, global demand for mineral resources will have to be multiplied by 4 by 2040. If we target a carbon neutral scenario in 2050, the demand will have to be multiplied by 6.

As the average grade of deposits of most metals tends to decrease, the number of mines will automatically increase in response of the increasing demand. Due to the presence of less exploited deposits and lower social and environmental standards, much of the mining is expected to shift to the global south (World Bank, 2020).
Yet, little is known about the consequences of mining in the South. Most of the literature on mining focuses on developed countries - Canada and Australia being two very important knowledge producers. 

Some metals are massively exploited only for the needs of the energy and digital transitions. Therefore, there's no historic.... Moreover, as the AIE illustrates, it takes on average 16,5  years for a mining project to start its production.  This means that there is no feedback on mining projects started after 2005.   we only have knowledge which reinforce this knowledge gap. If we look at some of the recent literature reviews on mining impacts, we find a majority of studies dealing with gold and coal mines.  

To adress this knowledge gap, we conducted a systematic review. A few systematic reviews have emerged in recent years linking mining and sustainable development. But none of them specifically address the impacts of transition metal mining on the Global South. This is despite the fact that most of the largest producers are located in the South and that this phenomenon is only expected to increase.

# Context
* **What are the energy transition minerals?**

The energy transition minerals are all the metals necessary for the construction of renewable energies and their carriers. This emerging typology is the result of renewable energy deployment projections. According to the International Energy Agency, demand for these metals is expected to surge in the coming years. The demand for these metals should indeed be multiplied by 4 by 2040 in order to meet the commitments of the Paris Agreement. If we want to reach net zero emissions in 2050, this demand would be multiplied by 6. In addition to the needs of the energy transition, we must add the needs of the digital transition and those of the rest of the economy. To meet this high demand, there will be no choice but to create new mines.  

* **The disproportionate place of gold, coal and other extractive industries in the litterature** 

If we look at the literature, we can see that researchers have mainly been interested in minerals with a long mining history, such as iron, gold or coal. This is normal since these resources have numerous extraction sites around the world, some of which have been exploited for decades. So there has been a lot of time to do studies.

One of the particularities of ETMs is the number of metals required for the needs of the energy transition. Some, such as lithium, have only been industrially mined for a few years and are still currently extracted in only a handful of extraction sites. These metals and their impacts have therefore been the subject of only a few recent studies. Even in recent systematic reviews we find mostly gold and coal mines (or at least, non transition mineral).
* **A word on the mining production structure (primary vs secondary metal)**

There is a large variety of energy transition minerals (cf: list below). This does not mean that there is mine (or several) for each mineral in the list.
There are primary metals, for which mine are created specifically for their extraction. There are also secondary metals which are entirely or partially extracted as by-products in the primary metal mines.

# Methodology

**1- Systematic review definition and approach**
* What's a (exploratory) systematic review?
* Following PRISMA guidelines
* SDGs as our main classification tool for our impacts


**2- Search strategy**

* For the search strategy, we have searched in Scopus, Web of Science and EBSCO host dabases.

* The search equation: we developed a search equation containing 3 parts : 
  * For the SDGs, we took advantage of the list of keywords proposed by Scopus. For each of the SDGs, Scopus provides a series of words adapted to each dimension of the 2030 Agenda. This allows us to target articles dealing with themes that can be linked to a goal without explicitly mentioning it. 
  * The second component of our search equation is a vector of keywords about the mine and the mining production system. We developed this list on the basis of our own personal knowledge but also . The objective was to have articles dealing with specific mines but also articles dealing with the mining industrial process as a whole. During the scoping phase, we realized that some articles did not specifically mention the word "mine" in the abstract and were therefore excluded from the Scopus and WoS searches. 
  * The third component of this equation is the list of countries called "Global South". It is a list of XX countries extracted from the World Bank report XXXXX. 
  * Finally, we also have excluded certain terms from the search. These are terms from the literature on "data mining" that frequently appeared in our results. We also excluded articles dealing with gold, oil and coal in their titles. During the scoping phase, articles containing one of these terms in their title were systematically out of scope. 

14 different searches with a unique equation each time were performed in the three academic databases. 

**3- Inclusion-exclusion criteria**

Given our research questions, we chose the following exclusion/inclusion criteria:
- Must establish an explicit link between the extraction of one or more TMEs and one or more economic or social impacts
- This impact must be determined by a clearly established method, whether quantitative or qualitative.
- Studies with a substansive value

**4- A mixed method review/design**

Integrated convergent analysis : Each impact, whether it comes from a quantitative or a qualitative analysis is coded the same way. Each impact is indeed classified as either positive, negative or neutral. 

The main advantage of this coding strategy, is that this puts qualitative and quantitatives studies on the same level. 

**5- Procedure of study classification for analysis**

For each selected article, we looked for sustainable development impacts. To be eligible, such impact must meet several criteria :
- It should be a clear causation going from the mine to a SD element
- It shoud be the result of a explicit methodology. As such, these sustainable development impacts are mostly extracted from the "result" part of the articles
- It should meet one of the SDG targets.

**6.1 Model**

Stanley, 2001 : *the independent variables often called moderator variables are those study characteristics that are thought to be consequential.*

\begin{align}
Y & = \beta_1 type\: of\: metal + \beta_2Field + \beta_3Precondition + \beta_4 Indirect \:effect + \beta_4 Mine\:type + \beta_5 Method\ \\
\end{align}

**Variable definitions:**


- **Y:** Dummy variable*. The outcome variable refers to the direction of sustainable impact (positive or negative). 

*Categorical variable in the multinomial probit estimation

- **Type of metal:** Vector of dummy variables. Designates the metal ore studied in the article.
- **Field:** Vector of dummy variables. Designates the academic field of the article
- **Precondition:** Vector of dummy variables. Refers to the socio-economic environment as described in the article.
- **Indirect effect:** Vector of dummy variables. Refers to the set of factors identified by the authors of the study as potentially influencing the nature of the mine's impact on sustainable development. 
- **Mine type:** Dummy variable. Indicates whether the mine is industrial or artisanal.
- **Method:**Dummy variable. Indicates whether  the article is a qualitative or a quantitative one.
